```python
import cv2
import numpy as np
from IPython.display import Image
```

# Moving average - 5 frames

```python
plik='GoodVision Traffic Counting using Camera Video Analytics _1.mp4'

cap = cv2.VideoCapture(plik)

fgbgAdaptiveGaussain = cv2.createBackgroundSubtractorMOG2()

frames_n = 5

frames_saved = []

while(1):
    ret, frame = cap.read()
    if not ret: break
        
    frames_saved.append(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))
    
    frames_to_mean = frames_saved[-frames_n:]
    frames_mean = np.mean(frames_to_mean, axis=0)
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frames_mean = cv2.GaussianBlur(frames_mean, (5, 5), 0).astype("uint8")
    fgmask = cv2.absdiff(frames_mean, gray_frame)
    _, fgmask = cv2.threshold(fgmask, 25, 255, cv2.THRESH_BINARY)

    fgbgAdaptiveGaussainmask = fgbgAdaptiveGaussain.apply(frame)
    cv2.namedWindow('Background Subtraction',0)
    cv2.namedWindow('Background Subtraction Adaptive Gaussian',0)
    cv2.namedWindow('Original',0)
    cv2.imshow('Background Subtraction',fgmask)
    cv2.imshow('Background Subtraction Adaptive Gaussian',fgbgAdaptiveGaussainmask)
    cv2.imshow('Original',frame)


    k = cv2.waitKey(1) & 0xff

    if k==ord('q'):
        break
cap.release()

cv2.destroyAllWindows()

print ('Program Closed')
```

# Moving average - 5 frames. Alpha -  0.1

```python
plik='GoodVision Traffic Counting using Camera Video Analytics _1.mp4'

cap = cv2.VideoCapture(plik)

fgbgAdaptiveGaussain = cv2.createBackgroundSubtractorMOG2()

_, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
initial_frame = np.float32(frame)

frames_saved = []

while(1):
    ret, frame = cap.read()
    if not ret: break
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.accumulateWeighted(gray_frame, initial_frame, 0.1)
    resultingFrames1 = cv2.convertScaleAbs(initial_frame)
    fgmask = cv2.absdiff(resultingFrames1, gray_frame)
    _, fgmask = cv2.threshold(fgmask, 25, 255, cv2.THRESH_BINARY)
    
    fgbgAdaptiveGaussainmask = fgbgAdaptiveGaussain.apply(frame)
    cv2.namedWindow('Background Subtraction',0)
    cv2.namedWindow('Background Subtraction Adaptive Gaussian',0)
    cv2.namedWindow('Original',0)
    
    cv2.imshow('Background Subtraction',fgmask)
    cv2.imshow('Background Subtraction Adaptive Gaussian',fgbgAdaptiveGaussainmask)
    cv2.imshow('Original',frame)


    k = cv2.waitKey(1) & 0xff

    if k==ord('q'):
        break
cap.release()

cv2.destroyAllWindows()

print ('Program Closed')
```

# Shadows removal, alfa = 0.2

```python
plik='GoodVision Traffic Counting using Camera Video Analytics _1.mp4'

cap = cv2.VideoCapture(plik)

fgbgAdaptiveGaussain = cv2.createBackgroundSubtractorMOG2()

_, frame = cap.read()
frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
initial_frame = np.float32(frame_gray)

frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
initial_frame_hsv = np.float32(frame_hsv)

alpha = 0
beta = 4
tauh = 255
taus = 5

while(1):
    ret, frame = cap.read()
    if not ret: break
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    hsv_mask = np.zeros_like(gray_frame)

    cv2.accumulateWeighted(gray_frame, initial_frame, 0.4)
    resultingFrames1 = cv2.convertScaleAbs(initial_frame)    
    
    cv2.accumulateWeighted(hsv_frame, initial_frame_hsv, 0.4)
    resultingFrames1_hsv = cv2.convertScaleAbs(initial_frame_hsv)
    resultingFrames1_hsv = initial_frame_hsv
    
    hsv_mask[np.where((hsv_frame[:,:,2]/resultingFrames1_hsv[:,:,2] <= beta) 
                      & (hsv_frame[:,:,2]/resultingFrames1_hsv[:,:,2] >= alpha) 
                      & (hsv_frame[:,:,1] - resultingFrames1_hsv[:,:,1] <= taus) 
                      & (np.abs(frame_hsv[:,:,0] - resultingFrames1_hsv[:,:,0]) <= tauh))] = 1
    hsv_mask[hsv_mask == 1] = 255
    
    fgmask = cv2.absdiff(resultingFrames1, gray_frame)
    _, fgmask = cv2.threshold(fgmask, 20, 255, cv2.THRESH_BINARY)
    
    fgbgAdaptiveGaussainmask = fgbgAdaptiveGaussain.apply(frame)
    cv2.namedWindow('Background Subtraction',0)
#     cv2.namedWindow('Background Subtraction Adaptive Gaussian',0)
#     cv2.namedWindow('Original',0)
    cv2.namedWindow('HSV',0)
    
    cv2.imshow('Background Subtraction', fgmask)
#     cv2.imshow('Background Subtraction Adaptive Gaussian', fgbgAdaptiveGaussainmask)
#     cv2.imshow('Original', frame)
    cv2.imshow('HSV', cv2.bitwise_and(hsv_mask, fgmask)) # 


    k = cv2.waitKey(1) & 0xff

    if k==ord('q'):
        break
cap.release()

cv2.destroyAllWindows()

print ('Program Closed')
```
